```python
import numpy as np
import pandas as pd
```

```python
df_test = pd.read_excel(r'Test_set.xlsx')
```

```python
df_test.head()
```

```python
df_test["Duration"].value_counts()
```

```python
df_test["Journey_day"] = pd.to_datetime(df_test.Date_of_Journey,format="%d/%m/%Y").dt.day
```

```python
df_test["Journey_month"] = pd.to_datetime(df_test.Date_of_Journey,format="%d/%m/%Y").dt.day
```

```python
df_test.drop(["Date_of_Journey"],axis=1,inplace=True)
```

```python
df_test.head()
```

```python
df_test["Dep_hour"] = pd.to_datetime(df_test["Dep_Time"]).dt.hour
df_test["Dep_mins"] = pd.to_datetime(df_test["Dep_Time"]).dt.minute

df_test.drop(["Dep_Time"],axis=1,inplace=True)
df_test.head()
```

```python
df_test["Arrival_hour"] = pd.to_datetime(df_test["Arrival_Time"]).dt.hour
df_test["Arrival_mins"] = pd.to_datetime(df_test["Arrival_Time"]).dt.minute
```

```python
df_test.drop(["Arrival_Time"],axis=1,inplace=True)
```

```python
df_test.head()
```

```python
duration = list(df_test["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"
        else:
            duration[i] = "0h "  + duration[i]
duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))
```

```python
df_test["Duration_hours"] = duration_hours
df_test["Duration_mins"] = duration_mins
```

```python
df_test.drop(["Duration"],axis=1,inplace=True)
```

```python
df_test.head()
```

```python
Airline = df_test[["Airline"]]

Airline = pd.get_dummies(Airline,drop_first=True)

Airline.head()
```

```python
Source = df_test[["Source"]]

Source = pd.get_dummies(Source,drop_first=True)

Source.head()
```

```python
Destination = df_test[["Destination"]]

Destination = pd.get_dummies(Destination,drop_first=True)

Destination.head()
```

```python
df_test.drop(["Route","Additional_Info"],axis=1,inplace=True)
df_test.head()
```

```python
df_test.replace({"non-stop":0, "1 stop":1, "2 stops":2, "3 stops":3, "4 stops":4},inplace=True)
```

```python
df_test.head()
```

```python
df_test_org = pd.concat([df_test,Airline,Source,Destination],axis=1)
```

```python
df_test_org.drop(["Airline","Source","Destination"],axis=1,inplace=True)
```
